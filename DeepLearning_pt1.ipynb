### Weights
Weights are numerical values that determine the importance of each input feature in a neural network.

- Every input is multiplied by a weight.
- Larger weight → more influence on the output
- Smaller (or negative) weight → less or inverse influence

### Bias
Bias is an additional parameter added to the weighted sum to shift the output.
 - Helps the model fit data better
 - Allows the model to represent relationships even when inputs are zero
 - Acts like an intercept in linear equations

### Neurons

A neuron in deep learning is a unit that takes inputs, multiplies them by weights, adds a bias, applies an activation function, and produces an output.

# Activation Functions in Deep Learning

## What is an Activation Function?
An activation function takes the output of a neuron and applies a transformation to introduce non-linearity, helping the model learn complex patterns.

---

## Common Activation Functions

### 1. ReLU (Rectified Linear Unit)
**Formula:**
$$
f(x) = \max(0, x)
$$

**Example:**
- Input: -3 → Output: 0  
- Input: 5 → Output: 5  

**Python Code:**
```python
import numpy as np

def relu(x):
    return np.maximum(0, x)

print(relu(-3))  # 0
print(relu(5))   # 5
```

## 2. Sigmoid Function

**Formula:**
$$
f(x) = \frac{1}{1 + e^{-x}}
$$

**Range:** (0, 1)

**Example:**
- Input: 0 → Output: 0.5  
- Input: 2 → Output: ~0.88

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

This document details the underlying structure and mathematics of a plain vanilla neural network, focusing on how information propagates through the system.

## 1. Network Architecture
At its core, a neural network is an incredibly complex mathematical function that takes numerical inputs and spits out numerical outputs.
*   **Neurons and Layers:** A neural network consists of layers of interconnected "neurons." A neuron is essentially just a container that holds a number, typically ranging between 0 and 1, which is referred to as its **activation**.
*   **Information Flow:** The network begins with an input layer, transitions through intermediate "hidden layers," and concludes with an output layer. The fundamental principle is that the activations in one layer strictly determine the activations in the next layer.
*   **Abstraction:** This layered structure allows the network to break down complex tasks into subproblems. It pieces together basic subcomponents in the early layers and combines them into increasingly abstract patterns and concepts in the later layers.

## 2. The Parameters: Weights and Biases
The specific way one layer influences the next relies on thousands of tunable parameters, which act as the dials and knobs that dictate the network's behavior.
*   **Weights:** The connections between neurons in consecutive layers are assigned **weights**, which are positive or negative numbers. A neuron in a subsequent layer computes a weighted sum of all the activations from the previous layer based on these connection weights.
*   **Biases:** To establish a threshold for when a neuron should actually light up, an additional number called a **bias** is added to the weighted sum. This tells the network how high the weighted sum needs to be before a neuron becomes meaningfully active.

## 3. The Feed Forward Process and Mathematical Explanation
The "feed forward" mechanism is the process by which activations transition from one layer to the next, ultimately transforming the initial input data into the final output prediction. To compute this transition efficiently, the network relies heavily on linear algebra.

We can represent the components of an entire layer mathematically:
*   Let **$\mathbf{a}$** be a vector organizing all the activations from the previous layer.
*   Let **$\mathbf{W}$** be a weight matrix, where each row corresponds to the connections between the previous layer and one specific neuron in the next layer.
*   Let **$\mathbf{b}$** be a vector containing all the biases for the next layer.

To feed the data forward, the network calculates the matrix-vector product of the weights and previous activations (**$\mathbf{W}\mathbf{a}$**). Then, the entire bias vector is added to that product (**$\mathbf{W}\mathbf{a} + \mathbf{b}$**).

Finally, to constrain the resulting values, this entire expression is wrapped in an activation function applied to each component of the vector. This gives us a neat, compact mathematical expression for the full transition of a layer:
$$ \mathbf{a}_{next} = \text{ActivationFunction}(\mathbf{W}\mathbf{a} + \mathbf{b}) $$
By iterating this matrix-vector multiplication and activation process across all layers, the network effectively feeds the input information forward to generate an output.

## 4. Activation Functions
The weighted sum and bias are pumped into an activation function to dictate the neuron's final output.

### Sigmoid Function
Historically, networks used the **sigmoid function** (also known as a logistic curve).
*   It squishes the real number line into a range between 0 and 1, forcing very negative inputs close to 0 and positive inputs close to 1.
*   Using a sigmoid, the feed-forward equation is written as: $$ \mathbf{a}_{next} = \sigma(\mathbf{W}\mathbf{a} + \mathbf{b}) $$.

### ReLU (Rectified Linear Unit)
Because sigmoids can make deep neural networks exceptionally difficult to train, modern networks overwhelmingly use the **ReLU** activation function.
*   ReLU simplifies the activation by taking the maximum of 0 and the input: $$ \text{ReLU}(x) = \max(0, x) $$.
*   With ReLU, if the input passes a certain zero threshold, the neuron acts as an identity function; if it does not, the neuron remains inactive and outputs zero.
