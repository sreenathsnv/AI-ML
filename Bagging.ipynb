### Is RF/DT are affected by outliers?

NO

 - Decision Trees and Random Forests are robust to outliers because:

 - They split using thresholds, not distances. ( Check entropy eqn)

 - They optimize class purity, not error magnitude.

 - Outliers usually end up in isolated leaf nodes.

 - Random Forest averages many trees, diluting outlier effects.
