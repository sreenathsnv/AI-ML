## Error

The indiviual difference from each predicted and actual value is called error ( y- y^)

## Cost Fuction
The Cumulative error or (MAE ) is called Cost function

image.png

```python
import numpy as np

y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

```python
def mae(y_true,y_predicted):

  total_error = 0
  for yt,yp in zip(y_true,y_predicted):
    total_error += abs(yt-yp)

  mae = total_error/len(y_true)
  return mae

```

```python
mae(y_true,y_predicted)
```

```python
def mse(y_true,y_predicted):

  total_error = 0
  for yt,yp in zip(y_true,y_predicted):
    total_error += (yt-yp)**2

  mae = total_error/len(y_true)
  return mae
```

```python
mse(y_true,y_predicted)
```

```python

def fix_log_input(arr,epsilon=1e-15):

  arr = [max(i,epsilon) for i in arr]
  arr = [min(i,1-epsilon) for i in arr]
  return arr

def cross_entropye(y_true,y_predicted):
  y_true =fix_log_input(y_true)
  y_predicted =fix_log_input(y_predicted)

  total_error = 0
  for yt,yp in zip(y_true,y_predicted):
    total_error += yt*np.log(yp) + (1-yt)*np.log(1-yp)
  return -(total_error/len(y_true))

```

```python
cross_entropye(y_true,y_predicted)
```
