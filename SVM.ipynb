# Support Vector Machine (SVM)

## Overview

Support Vector Machine (SVM) is a **powerful supervised learning algorithm** mainly used for **classification** and sometimes **regression**.

Unlike models such as **Logistic Regression**, which simply find a decision boundary, SVM:

- Finds the **best separating hyperplane**
- Creates **two marginal planes**
- Maximizes the **distance (margin)** between the classes

This improves **generalization and robustness**.

---

# 1. Hyperplane and Marginal Planes

The **decision boundary** in SVM is called a **hyperplane**.

Mathematically:

$$
W^T X + b = 0
$$

Where:

- \(W\) → weight vector  
- \(X\) → feature vector  
- \(b\) → bias term  

---

## Marginal Planes

Two parallel planes define the margin.

### Positive Marginal Plane

$$
W^T X + b = +1
$$

Points above this plane belong to:

$$
y = +1
$$

---

### Negative Marginal Plane

$$
W^T X + b = -1
$$

Points below this plane belong to:

$$
y = -1
$$

---

## Visualization

```
     +1 Class
       ●   ●
       │
WᵀX+b=1  (Positive Margin)
--------------------------
WᵀX+b=0  (Hyperplane)
--------------------------
WᵀX+b=-1 (Negative Margin)
       │
       ○   ○
      -1 Class
```

---

# 2. Margin Maximization

The **margin** is the distance between the two marginal planes.

Margin formula:

$$
Margin = \frac{2}{||W||}
$$

Where:

$$
||W|| = \sqrt{w_1^2 + w_2^2 + ... + w_n^2}
$$

### Optimization Goal

Maximize:

$$
\frac{2}{||W||}
$$

Equivalent to minimizing:

$$
\frac{1}{2}||W||^2
$$

---

# 3. Classification Constraint

To ensure correct classification:

$$
y_i (W^T X_i + b) \ge 1
$$

Where:

- \(y_i\) → class label (+1 or -1)  
- \(X_i\) → input feature  

This ensures:

- Positive points lie **above +1 margin**
- Negative points lie **below -1 margin**

---

# 4. Soft Margin (Handling Overlap)

Real datasets often contain **overlapping classes**.

SVM introduces **slack variables**.

Updated objective:

$$
\min \left(\frac{1}{2}||W||^2 + C \sum \eta_i \right)
$$

Where:

- **C** → penalty parameter controlling errors  
- **η (eta)** → slack variable measuring margin violation  

### Interpretation

| C Value | Behavior |
|------|------|
| Large C | Less tolerance for errors |
| Small C | Allows more margin violations |

---

# 5. Kernel Trick

When data is **not linearly separable**, SVM uses **kernels** to transform data into higher dimensions.

Common kernels:

| Kernel | Formula |
|------|------|
| Linear | \(X_i^T X_j\) |
| Polynomial | \((X_i^T X_j + 1)^d\) |
| RBF | \(e^{-\gamma||X_i-X_j||^2}\) |

# Mathematical Example

Dataset:

| Point | x1 | x2 | Class |
|-----|----|----|----|
| A | 1 | 2 | +1 |
| B | 2 | 3 | +1 |
| C | 3 | 1 | -1 |
| D | 4 | 2 | -1 |

Assume hyperplane:

$$
W = [1, -1]
$$

$$
b = 0
$$

---

### Step 1: Compute Hyperplane Output

Formula:

$$
W^T X + b
$$

For point A:

$$
(1)(1) + (-1)(2) + 0 = -1
$$

For point B:

$$
(1)(2) + (-1)(3) = -1
$$

For point C:

$$
(1)(3) + (-1)(1) = 2
$$

For point D:

$$
(1)(4) + (-1)(2) = 2
$$

---

### Step 2: Apply Classification Rule

$$
y_i (W^T X_i + b)
$$

For A:

$$
(+1)(-1) = -1
$$

Incorrect classification.

The optimization process will **adjust W and b** until:

$$
y_i (W^T X_i + b) \ge 1
$$

for all correctly classified points.

---

# Support Vectors

The points **closest to the margin** are called **support vectors**.

These points determine the **optimal hyperplane**.

```
●   (Support Vector)
      │
------Hyperplane------
      │
○   (Support Vector)
```

---

# Python Code Example

```python
import numpy as np
from sklearn import svm
import matplotlib.pyplot as plt

# Dataset
X = np.array([
    [1,2],
    [2,3],
    [3,1],
    [4,2]
])

y = np.array([1,1,-1,-1])

# Train SVM
model = svm.SVC(kernel='linear')

model.fit(X,y)

# Predictions
pred = model.predict(X)

print("Predictions:",pred)

# Support vectors
print("Support Vectors:")
print(model.support_vectors_)
```

# Advantages of SVM

- Works well with **high-dimensional data**
- Effective with **small datasets**
- Robust against **overfitting**

---

# Limitations

- Computationally expensive for **large datasets**
- Sensitive to **kernel choice**
- Hard to interpret compared to simpler models
